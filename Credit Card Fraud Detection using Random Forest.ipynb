# Invoking all the essential libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

## Loading Data to Pandas Dataframe

```python
credit_card_data=pd.read_csv('creditcard.csv.zip')
```

```python
#first 5 rows of dataset
credit_card_data.head()
```

```python
#last 5 rows of dataset
credit_card_data.tail()
```

## Dataset Information

```python
credit_card_data.info()
```

# Data Preprocessing

```python
#checking the number of missing values
credit_card_data.isnull().sum()
```

```python
#if any missing value is present ( although not present in this dataset)
credit_card_data.fillna(0, inplace=True)
```

# Distribution of legit transactions and fraudulent transactions

```python
credit_card_data['Class'].value_counts()
```

## This Data is highly unbalanced as legit transactions (0) are more than fraudulent transactions (1)

```python
#separating of data into two frames for analysis
legit=credit_card_data[credit_card_data.Class == 0]
fraud=credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

## Statistical measures of data 

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

# Now Comparing the values for both transactions

```python
credit_card_data.groupby('Class').mean()
```

### This shows a huge difference between the mean of legit transactions and fraudulent transactions

# Under Sampling 

### Build a new dataset containing similar distribution of normal transactions and fraudulent transaction.

```python
legit_sample=legit.sample(n=492)
```

### Concatenating two data frames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

### Dataset now has normal distribution of both transactions

```python
new_dataset.groupby('Class').mean()
```

# Splitting data into features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

# Split the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model training ( Random Forest Classifier)

```python
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train,Y_train)
```

# Model evaluation

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :', training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on training data :', test_data_accuracy)
```

```python
print(classification_report(X_test_prediction,Y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(X_test_prediction,Y_test))
```
